```python
# what percent of women are shorter than 154 cm
from scipy.stats import norm
height = 157
probability = norm.cdf(height, 161, 7)
print(f"{probability * 100:.2f}% of women in this survey are shorter than {height} cm")
```

```python
# persentage of the women taller than 154
height = 154
probability_taller = 1- norm.cdf(height, 161, 7)
print(f"{probability_taller * 100:.2f}% of women in this survey are taller than {height} cm")
```

```python
# between 154 cm and 157 cm the persentage
between= norm.cdf(157, 161, 7) - norm.cdf(154,161, 7)
print(between)
```

```python
import numpy as np

# Simulate rolling a 6-sided die 5 times with replacement
samp_5 = np.random.choice([1, 2, 3, 4, 5, 6], size=5, replace=True)
print(samp_5)
```

```python
np.mean(samp_5)
```

```python
# 10 defa tekrarlanan (Her 5 atisin ortalamsi)

sample_means = []
for i in range(10):
    samp_5 = np.random.choice([1, 2, 3, 4, 5, 6], size=5, replace=True)
    sample_means.append(float(np.mean(samp_5)))  # Convert np.float64 to float

print(sample_means)
```

```python
# Sampling Distributuin

sample_means =[]
for i in range(100000) :
    sample_means.append(float(np.mean(np.random.choice([1, 2, 3, 4, 5, 6], size=5, replace=True))))
    import matplotlib.pyplot as plt

plt.hist(sample_means)
plt.show()
```

```python
import pandas as pd

# Read the Excel file into a DataFrame
file_path = r"D:\python Folder\food_consumption.csv"  # Use raw string notation to handle backslashes
data = pd.read_csv(file_path)

# Display the first few rows of the DataFrame
print(data.head())
```

```python
data.shape
```

```python
#1. Plot the distribution of consumption for each food_category.
#Goal: Visualize how the consumption of different food categories is distributed.

#Question: How would you plot the distribution of food consumption for each food category?import seaborn as sns


import matplotlib.pyplot as plt

# Assuming 'data' is your DataFrame
sns.displot(data=data, y="consumption", hue="food_category", kind="kde")

# Display the plot
plt.show()
```

```python
# 2. 
# Create a bar plot showing the average co2_emission for each food_category.
# Goal: Compare the average CO2 emissions produced by each food category.

# Question: What type of Seaborn plot would you use to show the average CO2 emission by food category?

# Assuming 'data' is your DataFrame
sns.barplot(x="food_category", y="co2_emission", data=data, estimator="mean")

# Rotate x-axis labels for better readability if needed
plt.xticks(rotation=45)

# Add labels and title
plt.xlabel("Food Category")
plt.ylabel("Average CO2 Emission")
plt.title("Average CO2 Emission by Food Category")

# Display the plot
plt.show()
```

```python
sns.boxplot(x="food_category", y="consumption", data=data)

# Display the plot
plt.xticks(rotation=45)  # Rotate x labels for better readability
plt.show()
```
