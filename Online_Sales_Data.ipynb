```python
import matplotlib.pyplot as plt

# Veri hazÄ±rlÄ±ÄŸÄ±
profit_data = df.groupby(['Category', 'Sub-Category'])['Profit'].sum().round(2).sort_values()

# Ã‡ubuk grafiÄŸi oluÅŸtur
fig, ax = plt.subplots(figsize=(12, 6))

# Negatif deÄŸerler kÄ±rmÄ±zÄ±, pozitif deÄŸerler yeÅŸil renk olacak ÅŸekilde ayarla
colors = profit_data.apply(lambda x: 'green' if x > 0 else 'red')

# Ã‡ubuk grafiÄŸi Ã§iz
profit_data.plot(kind='bar', ax=ax, color=colors)

# BaÅŸlÄ±k ve etiketler
ax.set_title('Total Profit by Category and Sub-Category', fontsize=14)
ax.set_xlabel('Category and Sub-Category', fontsize=12)
ax.set_ylabel('Profit (in USD)', fontsize=12)

# Y ekseninde 1000 ayracÄ± ve yuvarlama
plt.tick_params(axis='y', labelsize=10)
plt.gca().get_yaxis().set_major_formatter(plt.FuncFormatter(lambda x, loc: "{:,}".format(x)))

# Grafik gÃ¶sterimi
plt.tight_layout()
plt.show()
```
