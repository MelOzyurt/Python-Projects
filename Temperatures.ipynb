```python
import pandas as pd
```

```python
# Path to the Excel file
file_path = r"D:\Python Folder\temperatures.csv"

# Read the Excel file
temperatures = pd.read_csv(file_path)
temperatures.head()
```

```python

# Convert the 'date' column to datetime format (if it's not already)
temperatures["date"] = pd.to_datetime(temperatures["date"])# Display the first few rows of the data
temperatures.head()

# Add a year column to temperatures
temperatures["year"] = temperatures["date"].dt.year
```

```python
# Pivot avg_temp_c by country and city vs year
temp_by_country_city_vs_year = temperatures.pivot_table("avg_temp_c", index=["country", "city"], columns="year").round(1)

# See the result
temp_by_country_city_vs_year
```

```python
# Subset for Egypt to India

temp_by_country_city_vs_year.loc["Egypt":"India", :]
```

```python
# Subset for Egypt, Cairo to India, Delhi
temp_by_country_city_vs_year.loc[("Egypt", "Cairo"):("India", "Delhi")]

```

```python
# Subset for Egypt, Cairo to India, Delhi, and 2005 to 2010
temp_by_country_city_vs_year.loc[("Egypt", "Cairo"):("India", "Delhi"), "2005":"2010"]
```

```python
temp_by_country_city_vs_year.head()
```

```python
# Calculate the mean temperature for each year, assigning to mean_temp_by_year

# Get the worldwide mean temp by year
mean_temp_by_year = temp_by_country_city_vs_year.mean()
mean_temp_by_year
```

```python
#Filter mean_temp_by_year for the year that had the highest mean temperature.
# Filter for the year that had the highest mean temp

print(mean_temp_by_year[mean_temp_by_year == mean_temp_by_year.max()])
```

```python
#Calculate the mean temperature for each city (across columns), assigning to mean_temp_by_city.
# Get the mean temp by city

mean_temp_by_city = temp_by_country_city_vs_year.mean(axis=1)
```

```python
mean_temp_by_city
```

```python

# Calculate the mean temperature for each city
mean_temp_by_city = temperatures.groupby('city')['avg_temp_c'].mean()

# Filter for the city that had the lowest mean temperature
city_with_lowest_temp = mean_temp_by_city[mean_temp_by_city == mean_temp_by_city.min()]

# Print the result
print(city_with_lowest_temp)
```
